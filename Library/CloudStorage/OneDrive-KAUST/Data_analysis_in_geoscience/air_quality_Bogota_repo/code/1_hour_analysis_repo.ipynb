```python
# plot actual data by hour
df_hour_actual = pd.DataFrame()
hours = np.arange(0, 24, 1)
for hour in hours:
    try:
        print(f"Processing hour: {hour}")
        data_by_hour = select_by_hour(selected_hour=hour, station_names=station_names, data_dic=data_dic)
        muni = gpd.sjoin(municipalities3, data_by_hour, how="left", op='contains').dropna()
        plot_actual_data_hourly(muni)
        # store PM2.5 by hour to plot after 
        temp_hour = muni.groupby(['DateTime', 'station_name'],as_index = False)['PM2.5'].mean()
        df_hour_actual = pd.concat([df_hour_actual, temp_hour])
        if hour != hours[-1]:
            plt.close('all')

    except Exception as e:
        print(f"An error occurred for hour {hour}: {e}")
        continue  # This will skip the rest of the current iteration and move to the next hour
```

```python
# plot the values of PM2.5 in Bogota
fig, ax = plt.subplots(1, 3,figsize=(15, 7))
labels = ['PM2.5 Levels [$\mu g/m3$]', 'PM10 Levels  [$\mu g/m3$]', 'CO Levels [ppm]']

# Plot values of PM2.5 for each municipality
muni_plot = muni.plot(column='PM2.5', ax=ax[0], legend=False, cmap='Blues')
# Plot the location of the stations
station_plot = data_by_hour.plot(ax=ax[0], column='station_name', legend=True)
# Create a colorbar for PM2.5
norm = Normalize(vmin=muni['PM2.5'].min(), vmax=muni['PM2.5'].max())
sm = ScalarMappable(norm=norm, cmap='Blues')
cbar = fig.colorbar(sm, ax=ax[0], shrink = 0.5)
cbar.set_label(labels[0])

# Plot values of PM10 for each municipality
muni_plot = muni.plot(column='PM10', ax=ax[1], legend=False, cmap='Blues')
# Plot the location of the stations
station_plot = data_by_hour.plot(ax=ax[1], column='station_name')
# Create a colorbar for PM10
norm = Normalize(vmin=muni['PM10'].min(), vmax=muni['PM10'].max())
sm = ScalarMappable(norm=norm, cmap='Blues')
cbar = fig.colorbar(sm, ax=ax[1], shrink = 0.5)
cbar.set_label(labels[1])

# Plot values of CO for each municipality
muni_plot = muni.plot(column='CO', ax=ax[2], legend=False, cmap='Blues')
# Plot the location of the stations
station_plot = data_by_hour.plot(ax=ax[2], column='station_name')
# Create a colorbar for CO
norm = Normalize(vmin=muni['CO'].min(), vmax=muni['CO'].max())
sm = ScalarMappable(norm=norm, cmap='Blues')
cbar = fig.colorbar(sm, ax=ax[2], shrink = 0.5)
cbar.set_label(labels[2])

#set the legend out of the plot
legend = ax[0].get_legend()
if legend:
    legend.set_bbox_to_anchor((-1.4,0.5))
    legend.set_loc('center left')
    legend.set_title('Municipality Stations')

ax[0].set_title(f"PM2.5 in Bogota during {data_by_hour['DateTime'].iloc[0]}", y = 1.02 )
ax[1].set_title(f"PM10 in Bogota during {data_by_hour['DateTime'].iloc[0]}", y = 1.02 )
ax[2].set_title(f"CO in Bogota during {data_by_hour['DateTime'].iloc[0]}", y = 1.02 )

ax[0].set_ylabel('Latitude')
ax[0].set_xlabel('Logitude')
ax[1].set_ylabel('Latitude')
ax[1].set_xlabel('Logitude')
ax[2].set_ylabel('Latitude')
ax[2].set_xlabel('Logitude')

plt.suptitle('Contamination Levels in Bogota', y = .83, fontsize = 15)
plt.tight_layout()
plt.show()
```

```python
# generate nearest neighborh interpolation by hour
df_hour_near = pd.DataFrame()
hours = np.arange(0, 24, 1)
for hour in hours:
    try:
        print(f"Processing hour: {hour}")
        data_by_hour = select_by_hour(selected_hour=hour, station_names=station_names, data_dic=data_dic)
        muni = gpd.sjoin(municipalities3, data_by_hour, how="left", op='contains').dropna()
        estimate_by_nearest_interpolator(muni, hourly_id= True)
        # store PM2.5 by hour to plot after 
        temp_hour = muni.groupby(['DateTime', 'station_name'],as_index = False)['PM2.5'].mean()
        df_hour_near = pd.concat([df_hour_near, temp_hour])
        if hour != hours[-1]:
            plt.close('all')

    except Exception as e:
        print(f"An error occurred for hour {hour}: {e}")
        continue  # This will skip the rest of the current iteration and move to the next hour
```

```python
animation_hourly_near = create_animation(folder_name = 'gif_make_hour', animation_name = 'hour_PM2.5.gif')
```

```python
#plot the time series data for the nearest estimation on the hourly analysis
fig, ax = plt.subplots(figsize = (30, 5))
sns.lineplot(data=df_hour_near, x="DateTime", y="PM2.5", hue= 'station_name', palette= 'colorblind')
ax.set_xticks(ax.get_xticks()[::2])
plt.xlabel('Hour')
plt.ylabel('PM2.5 Levels [$\mu g/m3$]')
plt.title('PM2.5 by Station in Bogota, [average by hour during 2023], Nearest Neighborh Estimation')
plt.show()
```
